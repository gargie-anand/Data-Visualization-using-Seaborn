```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
survey_data = pd.read_csv('young-people-survey-responses.csv')
```

```python
plot = sns.catplot(y = "Internet usage",
           data = survey_data,
           kind = "count")

plot.set_xticklabels(rotation = 0)
```

```python
plot = sns.catplot(y = "Internet usage",
           data = survey_data,
           kind = "count",
            col = "Age",
            col_order = ["21", "21 and older"])
```

It looks like most young people use the internet for a few hours every day, regardless of their age.

### Bar Plots with Percentages

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(x = "Gender", 
            y = "Mathematics",
           data = survey_data,
           kind = "bar")

plt.show()
```

This plot shows us that males report a much higher interest in math compared to females.

### Customizing bar plots

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
```

```python
# Create bar plot of average final grade in each study category
sns.catplot(x = "study_time", 
            y = "G3",
           data = student_data,
           kind = "bar")

plt.show()
```

```python
#list of catgories from lowest to highest
category_order = ["<2 hours",
                 "2 to 5 hours",
                 "5 to 10 hours",
                 ">10 hours"]

# Create bar plot of average final grade in each study category
sns.catplot(x = "study_time", 
            y = "G3",
           data = student_data,
           kind = "bar",
           order = category_order)

plt.show()
```

```python
#list of catgories from lowest to highest
category_order = ["<2 hours",
                 "2 to 5 hours",
                 "5 to 10 hours",
                 ">10 hours"]

# Create bar plot of average final grade in each study category
sns.catplot(x = "study_time", 
            y = "G3",
           data = student_data,
           kind = "bar",
           order = category_order,
           errorbar  = None)

plt.show()
```

Students in our sample who studied more have a slightly higher average grade, but it's not a strong relationship.
