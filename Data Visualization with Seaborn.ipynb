```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country = pd.read_csv('countries-of-the-world - Copy.csv')
```

### Making a Scatter Plot with a list

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x = "GDP", y = "Phones", data = country)
plt.show()
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x= "GDP", y = "percent_literate", data = country)
plt.show()
```

This plot does not show the linear relationship between GDP and percent literate, countries with the lower GDP do seem more likely to have a lower percent of the population that can read and write.

### Hue and Scatter Plots

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
```

```python
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location", hue_order =['Rural', 'Urban'])

# Show plot
plt.show()
```

It looks like students with higher absences tend to have lower grades in both rural and urban areas.

### Hue and Count Plots

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x = "school",
              data = student_data, 
              hue = "location", 
              palette = palette_colors)

# Display plot
plt.show()
```

Students at GP tend to come from an urban location, but students at MS are more evenly split

### Creating subplots with col and row

```python
# Change to use relplot() instead of scatterplot()
sns.relplot(x="absences", y="G3", 
                data=student_data, kind="scatter")

# Show plot
plt.show()
```

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", col="study_time")

# Show plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

### Creating two-factor subplots

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x = "G1", y = "G3", 
             data = student_data,
             kind = "scatter")



# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter",
            col="schoolsup",
            col_order=["yes", "no"])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row = "famsup",
            row_order=["yes", "no"])

# Show plot
plt.show()
```

### Changing the size of scatter plot points

```python
mpg = pd.read_csv('mpg.csv')
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x = "horsepower", y = "mpg", 
            data = mpg, kind = "scatter", size = "cylinders")



# Show plot
plt.show()
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders", 
            hue = "cylinders")

# Show plot
plt.show()
```

### Changing the style of scatter plot points

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x = "acceleration", y = "mpg", data = mpg,
            kind = "scatter", hue = "origin",
                style = "origin")



# Show plot
plt.show()
```

### Interpreting line plots

```python
# Create line plot
sns.relplot(x = "model_year", y = "mpg", data=mpg, kind = "line")


# Show plot
plt.show()
```

The shaded region represents a confidence interval for the mean, not the distribution of the observations.

### Visualizing standard deviation with line plots

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line",
            errorbar ="sd")

# Show plot
plt.show()
```

### Plotting subgroups in line plots

```python
# Create line plot of model year vs. horsepower
sns.relplot(x="model_year", y="horsepower",
            data=mpg, kind="line",
            errorbar = None)



# Show plot
plt.show()
```

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            errorbar=None,
            style="origin",
            hue="origin")

# Show plot
plt.show()
```

```python

# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            errorbar=None, style="origin", 
            hue="origin", markers = True, dashes = False)

# Show plot
plt.show()
```
